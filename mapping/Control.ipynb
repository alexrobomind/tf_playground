```python
import optuna
```

```python
optuna.delete_study(
    storage = 'sqlite:///storate.sqlite',
    study_name = 'streaming_1'
)
```

```python
study = optuna.create_study(
    storage = 'sqlite:///storate.sqlite',
    study_name = 'streaming_1',
    direction = 'minimize',
    sampler = optuna.samplers.TPESampler(),
    pruner = optuna.pruners.HyperbandPruner(),
    load_if_exists = True
)
```

```python
! optuna storage upgrade --storage sqlite:///storate.sqlite
```

```python
def set_range(name, min, max):
    study.set_user_attr('min_' + name, min)
    study.set_user_attr('max_' + name, max)
    
set_range('lr', 1e-3, 1e-2)
set_range('lr_decay', 0.1, 1)

set_range('n_layers', 1, 6)
set_range('d_layer', 32, 512)
set_range('n_cls', 10, 10)
```

```python
study.set_user_attr('baselines', ['new'])
```

```python
study.set_user_attr('t_max', 300)
study.set_user_attr('batch_size', 100000)
#study.set_user_attr('distance_file', 'md-distances-high_iota.nc')
#study.set_user_attr('pc_file', 'pc-high_iota.nc')
#study.set_user_attr('validation_split', 11)
study.set_user_attr('root', 'D:\\Daten\\OwnCloud Sciebo Juelich\\NNTrace\\Configs\\standard')
study.set_user_attr('n_phi', [0, 1, 5])
study.set_user_attr('center_loss_weight', 1e-5)
study.set_user_attr('min_scale', 4e-3)
#study.set_user_attr('baselines', ['studies\\tuning\\models\\39'])
```
